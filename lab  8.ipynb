```python
class Treenode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
#calculate the total number of nodes in the binary tree
    def totalnodes(self):
        if self is None:
            return 0
        return 1 + (self.left.totalnodes() if self.left else 0) + (self.right.totalnodes() if self.right else 0)
#count the number of leaf nodes in the binary tree
    def countleafnodes(self):
        if self is None:
            return 0
        if self.left is None and self.right is None:
            return 1
        return self.left.countleafnodes() + self.right.countleafnodes()

#calculate the sum of all nodes in the left subtree of the binary tree
    def sumleftSubTree(self):
        if self is None:
            return 0
        return self.data + (self.left.sumleftSubTree() if self.left else 0)
        

#count the number of nodes at level 'k' in the binary tree
    def countnodesSpecificlevel(self, k):
        if self is None:
            return 0
        if k == 0:
            return 1
        return (self.left.countnodesSpecificlevel(k - 1) if self.left else 0) + (self.right.countnodesSpecificlevel(k - 1) if self.right else 0)

#Count the number of full nodes (nodes with both left and right children) in the binary tree
    def countFullnodes(self):
        if self is None:
            return 0
        count = 0
        if self.left and self.right:
            count = 1
        return count + (self.left.countFullnodes() if self.left else 0) + (self.right.countFullnodes() if self.right else 0)

#insert a new node with data 'data' into the binary tree

    def insertnode(self, data):
        if data < self.data:
            if self.left is None:
                self.left = Treenode(data)
            else:
                self.left.insertnode(data)
        elif data > self.data:
            if self.right is None:
                self.right = Treenode(data)
            else:
                self.right.insertnode(data)


#perform post-order traversal of the binary tree

    def postOrder(self):
        if self:
            if self.left:
                self.left.postOrder()
            if self.right:
                self.right.postOrder()
            print(self.data, end=' ')

#Check whether the binary tree is a binary search tree (BST) or not

    def isBST(self, min_val=float('-inf'), max_val=float('inf')):
        if self is None:
            return True
        if self.data <= min_val or self.data >= max_val:
            return False
        return (self.left.isBST(min_val, self.data) if self.left else True) and (self.right.isBST(self.data, max_val) if self.right else True)
# Example usage:
root = Treenode(10)
root.insertnode(5)
root.insertnode(15)
root.insertnode(3)
root.insertnode(7)
root.insertnode(13)
root.insertnode(17)

print("total nodes:", root.totalnodes())
print("leaf nodes:", root.countleafnodes())
print("sum of nodes in left subtree:", root.sumleftSubTree())
print("nodes at level 2:", root.countnodesSpecificlevel(2))
print("full nodes:", root.countFullnodes())

root.insertnode(20)

print("Post-order traversal:")
root.postOrder()

print("\nIs it a BST?", root.isBST())
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.left = None
        self.right = None

class Order:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def total_price(self):
        return sum(product.price for product in self.products)

    def display_order(self):
        print("Ordered Products:")
        for product in self.products:
            print(f"{product.name}: ${product.price}")

class EcommerceSystem:
    def __init__(self, pin_code):
        self.root = None
        self.orders = []
        self.pin_code = pin_code

    def _validate_pin(self, pin_attempt):
        return pin_attempt == self.pin_code

    def _insert_product(self, node, product):
        if node is None:
            return product
        if product.price < node.price:
            node.left = self._insert_product(node.left, product)
        elif product.price > node.price:
            node.right = self._insert_product(node.right, product)
        return node

    def _search_product(self, node, price):
        if node is None or node.price == price:
            return node
        if price < node.price:
            return self._search_product(node.left, price)
        return self._search_product(node.right, price)

    def insert_product(self, name, price):
        new_product = Product(name, price)
        self.root = self._insert_product(self.root, new_product)

    def search_product_by_price(self, price):
        return self._search_product(self.root, price)

    def display_products_in_range(self, start_price, end_price):
        self._display_recursive(self.root, start_price, end_price)

    def _display_recursive(self, node, start_price, end_price):
        if node is None:
            return
        if start_price < node.price:
            self._display_recursive(node.left, start_price, end_price)
        if start_price <= node.price <= end_price:
            print(f"{node.name}: ${node.price}")
        if end_price > node.price:
            self._display_recursive(node.right, start_price, end_price)

    def place_order(self, order):
        self.orders.append(order)
        print("Order placed successfully!")

    def display_orders(self):
        print("All Orders:")
        for idx, order in enumerate(self.orders, start=1):
            print(f"Order {idx}:")
            order.display_order()

    def _manage_products(self, pin_attempt):
        if self._validate_pin(pin_attempt):
            print("Pin code accepted. Access granted to product management.")
            while True:
                print("Product Management Menu:")
                print("1. Add a new product")
                print("2. Exit product management")
                choice = input("Enter your choice: ")

                if choice == '1':
                    name = input("Enter the name of the new product: ")
                    price = float(input("Enter the price of the new product: "))
                    self.insert_product(name, price)
                elif choice == '2':
                    print("Exiting product management.")
                    break
                else:
                    print("Invalid choice. Please try again.")
        else:
            print("Incorrect pin code. Access denied to product management.")

# Function to display the menu options
def display_menu():
    print("Welcome to the E-commerce System!")
    print("1. Display all products")
    print("2. Search for a product by price")
    print("3. Display products within a price range")
    print("4. Place an order")
    print("5. Display all orders")
    print("6. Access Product Management")
    print("7. Exit")

# Example usage:
pin_code = "1234"  # Set your desired pin code
ecommerce_system = EcommerceSystem(pin_code)

# Inserting products
ecommerce_system.insert_product("Laptop", 1000)
ecommerce_system.insert_product("Smartphone", 800)
ecommerce_system.insert_product("Tablet", 500)
ecommerce_system.insert_product("Headphones", 200)

# Main program loop
while True:
    display_menu()
    choice = input("Enter your choice: ")

    if choice == '1':
        print("All products:")
        ecommerce_system.display_products_in_range(float("-inf"), float("inf"))
    elif choice == '2':
        price_to_search = float(input("Enter the price to search: "))
        found_product = ecommerce_system.search_product_by_price(price_to_search)
        if found_product:
            print(f"Product found: {found_product.name}: ${found_product.price}")
        else:
            print("Product not found.")
    elif choice == '3':
        start_price = float(input("Enter the start price: "))
        end_price = float(input("Enter the end price: "))
        print(f"Products within price range ${start_price} - ${end_price}:")
        ecommerce_system.display_products_in_range(start_price, end_price)
    elif choice == '4':
        order = Order()
        while True:
            product_price = float(input("Enter the price of the product to add (0 to finish): "))
            if product_price == 0:
                break
            product = ecommerce_system.search_product_by_price(product_price)
            if product:
                order.add_product(product)
            else:
                print("Product not found.")
        print(f"Total price of the order: ${order.total_price()}")
        ecommerce_system.place_order(order)
    elif choice == '5':
        ecommerce_system.display_orders()
    elif choice == '6':
        pin_attempt = input("Enter the pin code to access Product Management: ")
        ecommerce_system._manage_products(pin_attempt)
    elif choice == '7':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid choice. Please try again.")
```
